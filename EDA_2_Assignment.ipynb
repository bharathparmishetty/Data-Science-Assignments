```python
# Check for missing values represented as '?'
for col in df.columns:
  missing_count = df[col][df[col] == '?'].count()
  if missing_count > 0:
    print(f"Column '{col}' has {missing_count} missing values.")

# Replace '?' with NaN for easier handling
df.replace('?', np.nan, inplace=True)

# Impute missing values for numerical columns with the mean
numerical_cols = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

# Impute missing values for categorical columns with the mode
categorical_cols = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'income']
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Verify that there are no more missing values
print(df.isnull().sum())
```

```python
ss=StandardScaler()
df_ss=ss.fit_transform(df[numerical_cols])
pd.DataFrame(df_ss)
```

```python
mms=MinMaxScaler()
df_mms=mms.fit_transform(df[numerical_cols])
pd.DataFrame(df_mms)
```

```python
# •	Discuss the scenarios where each scaling technique is preferred and why.

# Scenarios for StandardScaler:
# 1. Algorithms that assume features are normally distributed: Many machine learning algorithms, like linear regression, logistic regression, and support vector machines, work best when features are normally distributed. StandardScaler transforms features to have zero mean and unit variance, which approximates a normal distribution.

# 2. Features with varying scales: When features have vastly different scales (e.g., age in years vs. income in dollars), StandardScaler ensures that all features contribute equally to the model.

# Scenarios for MinMaxScaler:
# 1. Algorithms sensitive to feature ranges: Some algorithms, like neural networks and k-nearest neighbors, are sensitive to the range of feature values. MinMaxScaler scales features to a specific range (usually 0 to 1), preventing features with larger ranges from dominating the model.

# 2. Preserving zero values: If zero values have a specific meaning in your data (e.g., zero income indicating unemployment), MinMaxScaler preserves these zeros while scaling other values.

# Why choose one over the other:
# - StandardScaler is generally preferred when you want to bring features to a similar scale without being overly concerned about outliers.
# - MinMaxScaler is preferred when you want to ensure features are within a specific range, especially when dealing with algorithms sensitive to feature ranges or when preserving zero values is important.

# The choice ultimately depends on your specific dataset, the algorithm you're using, and the desired properties of your scaled features.
```

Encoding Techniques

```python
# Identify columns for each encoding type
ohe_cols = [col for col in df.select_dtypes(include=["object"]) if df[col].nunique() < 5]
le_cols = [col for col in df.select_dtypes(include=["object"]) if df[col].nunique() >= 5]

# Apply One-Hot Encoding
ohe = OneHotEncoder(handle_unknown='ignore') # Handle potentially unseen categories in the future
ohe_df = pd.DataFrame(ohe.fit_transform(df[ohe_cols]).toarray(),
                      columns=ohe.get_feature_names_out(ohe_cols))

# Apply Label Encoding
le = LabelEncoder()
for col in le_cols:
    df[col] = le.fit_transform(df[col])

# Combine the results
final_df = pd.concat([df.drop(ohe_cols, axis=1), ohe_df], axis=1)
```

```python
#Discuss the pros and cons of One-Hot Encoding and Label Encoding.

# Pros and Cons of One-Hot Encoding:

# Pros:
# 1. Handles Nominal Data: One-Hot Encoding is ideal for nominal categorical features (categories with no inherent order) as it creates binary features for each category, avoiding the creation of artificial ordinal relationships.

# 2. Improved Model Performance: By representing categories as independent features, One-Hot Encoding can improve the performance of many machine learning algorithms, especially those sensitive to feature scaling and ordinal relationships.

# Cons:
# 1. Increased Dimensionality: For features with many unique categories, One-Hot Encoding can significantly increase the number of features, leading to potential computational costs and overfitting.

# 2. Sparsity: The resulting one-hot encoded features are often sparse (mostly zeros), which might require specialized algorithms or data structures for efficient handling.

# Pros and Cons of Label Encoding:

# Pros:
# 1. Simplicity and Efficiency: Label Encoding is straightforward and computationally efficient, especially for features with a large number of unique categories.

# 2. Preserves Ordinal Relationships: For ordinal categorical features (categories with a natural order), Label Encoding maintains the inherent order, which can be beneficial for some algorithms.

# Cons:
# 1. Creates Artificial Ordinal Relationships: For nominal features, Label Encoding introduces artificial ordinal relationships between categories, which can mislead some algorithms and negatively impact performance.

# 2. Limited Applicability: Label Encoding is generally not suitable for algorithms sensitive to feature scaling or those that assume features are normally distributed.
```

Feature Engineering

```python
#Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.

# Feature 1: Combined Capital
# Rationale: Combining capital gain and loss into a single feature might provide a more holistic view of an individual's financial status.
final_df['combined_capital'] = final_df['capital_gain'] - final_df['capital_loss']

# Feature 2: Hours Worked Per Week Category
# Rationale: Categorizing hours worked per week could capture potential non-linear relationships between working hours and income.
bins = [0, 30, 40, 60, 100]
labels = ['Part-time', 'Full-time', 'Overtime', 'Extensive']
final_df['hours_per_week_category'] = pd.cut(final_df['hours_per_week'], bins=bins, labels=labels)

# Apply Label Encoding to the new categorical feature
le = LabelEncoder()
final_df['hours_per_week_category'] = le.fit_transform(final_df['hours_per_week_category'])

final_df.head()
```

```python
#Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.

# Check for skewness in numerical features
skewed_features = final_df[numerical_cols].skew()
print(skewed_features)

# Apply log transformation to 'capital_gain' due to high skewness
final_df['capital_gain_log'] = np.log1p(final_df['capital_gain'])  # Use log1p to handle zero values

# Justification:
# The 'capital_gain' feature exhibits high skewness, indicating a long tail of high-income earners.
# Log transformation compresses the range of values, reducing the influence of extreme outliers and potentially improving model performance.
# It can help normalize the distribution and make it more suitable for algorithms that assume normality.
```

```python
from sklearn.ensemble import IsolationForest

# Create an IsolationForest model
iso_forest = IsolationForest(contamination=0.05)  # Adjust contamination as needed

# Fit the model and predict outliers
outliers = iso_forest.fit_predict(final_df.select_dtypes(include=['number']))

# Remove outliers from the DataFrame
final_df_no_outliers = final_df[outliers != -1]
final_df_no_outliers
```

```python

# Outliers can significantly affect model performance in various ways:

# 1. Bias in Parameter Estimation:
#    - Outliers can disproportionately influence the estimation of model parameters, pulling them away from values that represent the majority of the data.
#    - This can lead to biased models that make inaccurate predictions, especially for new data points that fall within the normal range.

# 2. Increased Variance and Overfitting:
#    - Outliers can increase the variance of a model, making it more sensitive to small fluctuations in the training data.
#    - This can lead to overfitting, where the model learns the training data too well, including the outliers, and performs poorly on unseen data.

# 3. Reduced Model Robustness:
#    - Models trained on data with outliers may become less robust and more susceptible to the influence of noisy or extreme data points in the future.
#    - This can decrease the model's generalization ability and reliability in real-world scenarios.

# 4. Misinterpretation of Relationships:
#    - Outliers can distort the perceived relationships between variables in the data.
#    - This can lead to incorrect conclusions about the underlying patterns and dependencies, hindering the development of accurate and insightful models.

# 5. Impact on Specific Algorithms:
#    - Certain algorithms are more sensitive to outliers than others. For example:
#        - Linear regression models can be heavily influenced by outliers due to their reliance on minimizing squared errors.
#        - K-means clustering can be affected by outliers as they can distort the cluster centroids.

# Mitigation Strategies:
#    - Outlier detection and removal: Identify and remove or replace outliers before training the model.
#    - Robust algorithms: Use algorithms that are less sensitive to outliers, such as robust regression or decision trees.
#    - Data transformation: Apply transformations like log or square root to reduce the impact of extreme values.
#    - Careful feature engineering: Design features that are less susceptible to outliers or capture their information in a meaningful way.
```

```python
!pip install ppscore
import ppscore as pps

# Calculate the Predictive Power Score (PPS) matrix
pps_matrix = pps.matrix(final_df_no_outliers)

# Display the PPS matrix
pd.DataFrame(pps_matrix)
```

```python

# Calculate the correlation matrix
corr_matrix = final_df_no_outliers.corr()

# Display the correlation matrix
pd.DataFrame(corr_matrix)

# Comparing PPS and Correlation:

# Predictive Power Score (PPS):
# - Measures the ability of one feature to predict another feature.
# - Captures both linear and non-linear relationships.
# - Provides a directional measure (predictor -> predicted).
# - Ranges from 0 (no predictive power) to 1 (perfect prediction).

# Correlation:
# - Measures the linear association between two features.
# - Does not capture non-linear relationships.
# - Provides a symmetrical measure (no directionality).
# - Ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation).

# Key Differences and Insights:

# - Non-linear Relationships: PPS can reveal predictive relationships that correlation might miss due to its focus on linear associations.
# - Directionality: PPS indicates the direction of the predictive relationship, while correlation only shows the strength of association.
# - Complementary Analysis: Using both PPS and correlation can provide a more comprehensive understanding of feature relationships.

# Example:
# - If PPS shows a high score for feature A predicting feature B, but correlation is low, it suggests a non-linear relationship.
# - If both PPS and correlation are high, it reinforces a strong linear relationship.

# By comparing the PPS matrix and correlation matrix, you can gain deeper insights into the complex relationships between features in your dataset, potentially leading to better feature selection and model development.
```

```python
corr_abs=corr_matrix.abs()
np.fill_diagonal(corr_abs.values,0)
best_corr_value=corr_abs.max().max()
best_corr_value
```

```python
best_corr_indices=np.where(corr_abs==best_corr_value)
best_corr_pair=[(corr_matrix.index[i],corr_matrix.columns[j]) for i,j in zip(*best_corr_indices)]
best_corr_pair
```
